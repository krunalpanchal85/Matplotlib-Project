```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
citi_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

citi_data_df= citi_data_df.drop_duplicates()
```

```python
# Combine the data into a single dataset
combined_citiride_df = pd.merge(citi_data_df, ride_data_df, how='outer', on='city')
combined_citiride_df= combined_citiride_df.drop_duplicates()
# Display the data table for preview
combined_citiride_df
```

## Bubble Plot of Ride Sharing Data

```python
grouped_citiride_df = combined_citiride_df.groupby(['type']+['city'])

#Rides per city
rides_per_city = grouped_citiride_df['ride_id'].count()
urban_rides_per_city = rides_per_city['Urban']
suburban_rides_per_city = rides_per_city['Suburban']
rural_rides_per_city = rides_per_city['Rural']

#Average fare
Avg_fare = grouped_citiride_df['fare'].mean()
urban_avg_fare = Avg_fare['Urban']
rural_avg_fare = Avg_fare['Rural']
Suburban_avg_fare = Avg_fare['Suburban']

#Driver Count
driver_count = grouped_citiride_df['driver_count'].max()

driver_count
rural_driver_count = driver_count['Rural']
suburban_driver_count = driver_count['Suburban']
urban_driver_count = driver_count['Urban']

facecolors=driver_count
# Build the scatter plots for each city types
urban = plt.scatter(urban_rides_per_city, urban_avg_fare, marker="o", facecolors="coral", 
                    edgecolors="black",s= 10*urban_driver_count, alpha= '0.75', label = 'Urban')
suburban =plt.scatter(suburban_rides_per_city, Suburban_avg_fare, marker="o", facecolors="skyblue", 
                      edgecolors="black",s=10*suburban_driver_count, alpha='0.75',label='Suburban')
rural= plt.scatter(rural_rides_per_city, rural_avg_fare, marker="o", facecolors="gold", 
                   edgecolors="black",s=10*rural_driver_count, alpha='0.75',label='Rural')

# Create a legend
#plt.legend(title='City Types',loc='best')
legend = plt.legend(title="City Types", loc='best', prop={'size': 8})
legend.legendHandles[0]._sizes = [100]
legend.legendHandles[1]._sizes = [100]
legend.legendHandles[2]._sizes = [100]
plt.grid()
# Create a title, x label, and y label for our chart
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")


# Save Figure
plt.savefig("Pyber_Ride_Sharing.png")

plt.show()
```

## Total Fares by City Type

```python
grouped_citiride_df = combined_citiride_df.groupby(['type'])


#Average fare
Avg_fare_df = (grouped_citiride_df['fare'].sum()/combined_citiride_df['fare'].sum())*100
Avg_fare_df

pie_state_type = ['Rural','Suburban','Urban']
colors = ["gold","skyblue","coral"]
explode = (0, 0, 0.1)

# Tell matplotlib to create a pie chart based upon the above data
plt.pie(Avg_fare_df, explode=explode, labels=pie_state_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=165)

#title
plt.title("% of Total Fares by City Type")
# Create axes which are equal so we have a perfect circle
plt.axis("off")

# Save an image of our chart and print the final product to the screen
plt.savefig("Total_Fares_by_City_Type.png")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Build Pie Chart

# Save Figure
grouped_citiride_df = combined_citiride_df.groupby(['type'])


#Average fare
Avg_ride_df = (grouped_citiride_df['ride_id'].count()/combined_citiride_df['ride_id'].count())*100
Avg_ride_df

grouped_citiride_df['ride_id'].count()
combined_citiride_df['ride_id'].count()
pie_state_type = ['Rural','Suburban','Urban']
colors = ["gold","skyblue","coral"]
explode = (0, 0, 0.1)

# Tell matplotlib to create a pie chart based upon the above data
plt.pie(Avg_ride_df, explode=explode, labels=pie_state_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=165, pctdistance = 0.7)

#title
plt.title("% of Total Rides by City Type")
# Create axes which are equal so we have a perfect circle
plt.axis("image")

# Save an image of our chart and print the final product to the screen
plt.savefig("Total_Rides_by_City_Type.png")
```

## Total Drivers by City Type

```python
# Build Pie Charts

# Save Figure
grouped_citiride_df = citi_data_df.groupby(['type'])


Avg_driver_df = (grouped_citiride_df['driver_count'].sum()/citi_data_df['driver_count'].sum())*100

pie_state_type = ['Rural','Suburban','Urban']
colors = ["gold","skyblue","coral"]
explode = (0, 0, 0.1)

# Tell matplotlib to create a pie chart based upon the above data
plt.pie(Avg_driver_df, explode=explode, labels=pie_state_type, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=165)

#title
plt.title("% of Total Drivers by City Type")
# Create axes which are equal so we have a perfect circle
plt.axis("equal")

# Save an image of our chart and print the final product to the screen
plt.savefig("Total_Driver_by_City_Type.png")

plt.show()
```
